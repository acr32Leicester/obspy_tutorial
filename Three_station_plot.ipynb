```python
import datetime
import numpy as np
from obspy import UTCDateTime, read_inventory, geodetics
#
from obspy.clients.fdsn import Client
#
import matplotlib.pyplot as plt
#import matplotlib.dates as dates
#import matplotlib.gridspec as gridspec
#from matplotlib.ticker import FuncFormatter as fnfrm


# Station map location: http://ds.iris.edu/gmap

ev_time="2021-06-05 17:55:58"
ev_mag = 6.6
#ev_depth = 10
#ev_lat = 0.95
#ev_long = 21.725

#Location = "Leicester, UK"
STATION = "RE9A7"
NETWORK= 'AM'
LOC_CODE = '00'
CHANNEL = 'EHZ'
CLIENT = "RASPISHAKE"


client = Client(CLIENT)

ev_start = UTCDateTime(ev_time)
trim_value = 5
trim_start = ev_start - trim_value
trim_end = ev_start + 10*60 + trim_value

#inv = client.get_stations(network="CI", station="R*",starttime=trim_start, endtime=trim_end, level="channel")
#print(inv)

#inv.plot()

#inv = read_inventory()

st1 = client.get_waveforms(NETWORK, STATION, LOC_CODE, CHANNEL, trim_start, trim_end, attach_response=True)



pre_filt=(0.05, 0.1, 45.0, 50.0)
#tr =st1[0]
st1 = st1.remove_response(pre_filt=pre_filt, output='VEL')
st1.detrend('spline',order=4, dspline=500, plot=False)


st1.filter('bandpass', freqmin=0.1, freqmax=1.0, corners=4, zerophase=True)
st1.trim(starttime=trim_start + trim_value, endtime=trim_end - trim_value)
st1 = st1[0]
st1.times("relative")
print(st1.times("relative"))

plt.plot(st1.times("relative"), st1.data)
plt.show()

# Setting up traces and converting from counts to Velocity
# pre_filt=(0.05, 0.1, 45.0, 50.0)
# st1 = st1.remove_response(inventory=inv, pre_filt=pre_filt, output='VEL')
# st1.detrend('spline',order=4, dspline=500, plot=False)
# label= []
# wave_col = []
```
